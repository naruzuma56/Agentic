# Messages
Messages are the fundatmental unit of context for models in LangChain. They represent the input and output of models, carruing both the content and meta data needed to represent the stat of a converstaion when interacting with an LLM. Messages are objects that contain:
--- 
- * Role:- Identifies the message type(e.g. system, user)
---
- * Content:- Represents the actual content of the message(lice : text images, audio docs, etc.)
---
- * Metadate:- Optional fields such as response  information, message IDs and tocen usage/
- Langchain provides a standard message type that worcs across all model providers, ensuring consistent behaviour regardless of the model being called.

```python
import os
from langchain_openrouter import ChatOpenRouter
os.environ["OPENROUTER_API_KEY"] = os.getenv("OPENROUTER_API_KEY")

model = ChatOpenRouter(model="openrouter/owl-alpha")
model
```

```python
model.invoke("Please tell me what is ai") # by default treated as human message
```

## Text Prompts
Text prompts are strings - ideal for straightforward geneartion tascs where you don't need to retain conversation history.
* Use text prompts when:
1. You have a single, standalone request
2. You don't need converstaion history
3. You want minimal code complexity

## Message Prompts 
Alternatively, you can pass in a list of messages to the model by providing a list of message objects.
* Messages types
1. System Messages: A System message represents an intiial set of instructions that primes the models's behaviour. You can use a system message to set the tone, define the models's role and establish guidlines for responses.
2. Human message: A Human Message represents user input and interactions. They can contain text, images , audios, files and any other amout of multimodal content.
3. AI message: An AI message represents the output of a model invocation. They can include multimodal data, tool calls and provider-specific metadata that you can later access
4. Tool Message: For models that support tool calling, AI messages can contain tool calls. Tool messages are used to pass the results of a single tool execution bacc to the model.


```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage

messages=[
    SystemMessage("YOu are a poertry expert"),
    HumanMessage("Write a poem on AI")
]
response =model.invoke(messages)
response.content
```

```python
system_mes = SystemMessage("YOu are a helpful conding assistant") 
messages=[
    system_mes,
    HumanMessage("How do I create a rest api")
]
response =model.invoke(messages)
print(response.content)
```

```python
## Detailed info to the LLM thorugh system message

system_mes = SystemMessage("""
You are s senior Pyhton developer with expertise in web frameworks.
Always provide code examples and explain you reasoning.
Be concise but thorough in you explanation.
""")

messages = [
    system_mes,
    HumanMessage("How do i create rest api")
]
response = model.invoke(messages)
print(response.content)
```

```python
### Message metadata
human_msg = HumanMessage(
    content="Hello",
    name="ALice",
    id="msg_123",
)

response = model.invoke([
    human_msg
])
response
```

```python
from langchain.messages import AIMessage, SystemMessage, HumanMessage

ai_msg = AIMessage("I'd be happt ot help you with that question")

messages = [
    SystemMessage("You are a helpful assistant"),
    HumanMessage("Can you help me?"),
    ai_msg,
    HumanMessage("Great , what 3456 * 325423")
]
response = model.invoke(messages)
print(response.content)
```
